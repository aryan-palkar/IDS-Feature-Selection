```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('store/result3-39_multi34.csv')

df[['accuracy_score', 'recall_score', 'precision_score', 'f1_score']] = df[['accuracy_score', 'recall_score', 'precision_score', 'f1_score']] * 100
df[['accuracy_score', 'recall_score', 'precision_score', 'f1_score']] = df[['accuracy_score', 'recall_score', 'precision_score', 'f1_score']].round(2)

df.head(10)
```

```python
NO_OF_MODELS = 3
NO_OF_FEATURESETS = 111

newDF = pd.DataFrame()

for i in range(0, NO_OF_MODELS):
    MODEL_NAME = df['model'][i]
    resullts = []
    print(MODEL_NAME)
    for j in range(i, NO_OF_FEATURESETS, NO_OF_MODELS):
        resullts.append(df['accuracy_score'][j])
    newDF[MODEL_NAME] = resullts

# print(newDF)

newDF[['RandomForestClassifier','DecisionTreeClassifier']].plot()
plt.show()
```
